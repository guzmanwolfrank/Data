```python
import pandas as pd

filepath = r'https://raw.githubusercontent.com/guzmanwolfrank/Data-SQL/TradeSQL/TradeSQL/wolfranktrades%20-%20tradesOutput.csv'

df = pd.read_csv(filepath)

df['Account'] = 'Apex Block Account'

# Remove all rows with NaN or null values
df.dropna(inplace=True)
print("Removed all rows with NaN or null values.")

# Rename cum. net profit column to Cumprofit
df = df.rename(columns={'Cum. net profit': 'Cumprofit'})

# Convert currency columns to integer
df['Cumprofit'] = df['Cumprofit'].replace('[\$,]', '', regex=True).astype(float).astype(int)
df['Profit'] = df['Profit'].replace('[\$,]', '', regex=True).astype(float).astype(int)
df['Commission'] = df['Commission'].replace('[\$,]', '', regex=True).astype(float).astype(int)

# Calculate Net Profit
df['NetProfit'] = df['Profit'] - df['Commission']

# Define initial balance
initial_balance = 10000

# Calculate Balance
df['Balance'] = df['NetProfit'].cumsum() + initial_balance

# Calculate Return
df['Return'] = df['NetProfit'] / df['Balance']

# Display the DataFrame
display(df)
```
