```python
import csv
import random
import string
from faker import Faker
import os
import pandas as pd

# Initialize Faker
fake = Faker()

# Define the number of records to generate
num_records = 100

# Define possible values for certain fields
transaction_types = ['WITHDRAWAL', 'DEPOSIT', 'TRANSFER', 'REFUND', 'REBATE']
card_types = ['Virtual', 'Physical']
card_states = ['ACTIVE', 'TERMINATED', 'SUSPENDED']
channels = ['ADMIN', 'FRAUD', 'API', 'IVR', 'SYSTEM']
mcc_group_names = ['BROKERAGE', 'BANK', 'RETAIL', 'ECOMMERCE', 'DELIVERY']
nationalities = ['jp_JP', 'en_US', 'es_MX', 'ru_RU', 'hi_IN']
gps_ids = ['GPS001', 'GPS002', 'GPS003', 'GPS004', 'GPS005']

# Define the output CSV file path
output_file = 'fake_banking_data.csv'

# Open a CSV file for writing
with open(output_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # Write the header row
    writer.writerow([
        'TransactionID', 'AccountID', 'TransactionDate', 'Amount', 'TransactionType', 
        'Description', 'First Name', 'Last Name', 'VendorID', 'FeeID', 'FeePayable', 
        'Card', 'MCC GroupName', 'Channel', 'CardState', 'CardToken', 
        'Server_Location', 'User_Location'
    ])
    
    # Write the fake data rows
    for _ in range(num_records):
        transaction_id = fake.uuid4()
        account_id = fake.uuid4()
        transaction_date = fake.date_this_year()
        amount = round(random.uniform(10.0, 1000.0), 2)
        transaction_type = random.choice(transaction_types)
        description = fake.sentence()
        nationality = random.choice(nationalities)
        
        # Generate first and last names based on nationality
        fake.seed_locale(nationality)
        first_name = fake.first_name()
        last_name = fake.last_name()
        
        vendor_id = fake.uuid4()
        fee_id = fake.uuid4()
        fee_payable = round(random.uniform(1.0, 50.0), 2)
        card = random.choice(card_types)
        mcc_group_name = random.choice(mcc_group_names)
        channel = random.choice(channels)
        card_state = random.choice(card_states)
        
        # Generate a random card token with 15 characters of letters and numbers, capitalized and lowercase
        card_token = ''.join(random.choices(string.ascii_letters + string.digits, k=15)).upper()
        
        # Set Server_Location and User_Location equal to a GPS ID
        server_location = random.choice(gps_ids)
        user_location = random.choice(gps_ids)

        writer.writerow([
            transaction_id, account_id, transaction_date, amount, transaction_type, 
            description, first_name, last_name, vendor_id, fee_id, fee_payable, 
            card, mcc_group_name, channel, card_state, card_token, 
            server_location, user_location
        ])

# Get the absolute path of the CSV file
file_path = os.path.abspath(output_file)
print(f"Fake data generated and saved to '{file_path}'.")

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# Display the DataFrame
print(df.head())

```
