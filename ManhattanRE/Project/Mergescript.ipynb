# Merge Script for Data 

## After creating the longitude and lattitude coordinates from zipcodes using the Nominatim module, we can now merge the Manhattan Data spreadsheet with the manzip_coordinates spreadsheet using dataframes.  Once we clean the new merged data with dropna inplace method, we can now use this new spreadsheet data to create Geographical data on dashboards and other visualization devices. 



```python
import pandas as pd
from geopy.geocoders import Nominatim
import requests 
```

```python


url1 = "https://raw.githubusercontent.com/guzmanwolfrank/Data-SQL/main/ManhattanRE/Data/ManhattanData.csv"  # Replace with the actual URL
data = pd.read_csv(url1)


url2 = "https://raw.githubusercontent.com/guzmanwolfrank/Data-SQL/main/ManhattanRE/Data/manzip_coordinates.csv"  # Replace with the actual URL
zip = pd.read_csv(url2)


display (data)
display (zip)
```

```python

# Merge the 'data' DataFrame with 'xyz_data' based on 'zipcode'
merged_data = pd.merge(data, zip, on="ZIP_CODE", how="outer")

ndata= merged_data 

# Let's clean the data 
ndata.dropna(inplace=True)

# Display the new, clean merged data.  It is now ready for geographical data analysis and other visualizations like dashboards! 
display(ndata)
```

```python
ndata.to_csv("Data\Merged.csv", index=True)
```
